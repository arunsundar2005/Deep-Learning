```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

x_data = np.random.rand(100).astype(np.float32)
y_data = x_data * 0.1 + 0.3


W = tf.Variable(tf.random.normal([1]))
b = tf.Variable(tf.zeros([1]))

print(x_data)
print(y_data)
print(W)
print(b)


def mse_loss():
  y = W*x_data+b
  loss = tf.reduce_mean(tf.square(y-y_data))
  return loss

optimizer = keras.optimizers.Adam()

for step in range(5000):
  optimizer.minimize(mse_loss,var_list=[W,b])
  if step % 50 == 0:
    print(step,W.numpy(),b.numpy())

print("Final Data : ",W.numpy(),b.numpy() )
```

```python
import numpy as np
import tensorflow as tf

N=20

polynomial = np.poly1d([1,2,3])
print("polynomial is ", polynomial)
X = np.random.uniform(-10,10, size = (N,1))
Y = polynomial(X)
print("Y is", Y)

fx = np.hstack([X*X, X, np.ones_like(X)])
print("fx is ", format(fx))

w = tf.Variable(tf.random.normal((3,1)))
x = tf.constant(fx, dtype = tf.float32)
print("x is ", x)
y = tf.constant(Y, dtype = tf.float32)

optimizer = tf.keras.optimizers.Nadam(learning_rate=0.01)
print(w)

for _ in range(10):
  with tf.GradientTape() as tape:
    y_pred = x @ w
    print("preedicted :", y_pred)
    mse = tf.reduce_mean(tf.square(y_pred-y))
  grad = tape.gradient(mse, w)
  optimizer.apply_gradients([(grad, w)])

print(w)
```

```python
# Solving a math problem
# A + B = 8
# +   +
# C - D = 6
#||  ||
# 13  8

import tensorflow as tf
import numpy as np
import random
from tqdm import tqdm

A = tf.Variable(random.random())
B = tf.Variable(random.random())
C = tf.Variable(random.random())
D = tf.Variable(random.random())


Epochs =5000
optimizer = tf.keras.optimizers.Nadam(learning_rate = 0.01)

for _ in tqdm(range(Epochs), desc = "Training Progress", ncols=100, bar_format="{l_bar}{bar} | {n_fmt}/{total_fmt} [{elapsed}<{remaining}]"):
  with tf.GradientTape() as tape:
    y1 = A + B -8
    y2 = C - D - 6
    y3 = A + C - 13
    y4 = B + D - 8
    loss = y1*y1 + y2*y2 + y3*y3 + y4*y4
  gradA, gradB, gradC, gradD = tape.gradient(loss, [A,B,C,D])
  optimizer.apply_gradients([(gradA, A), (gradB, B), (gradC, C), (gradD, D)])


print("\nA :" , A.numpy())
print("B :" , B.numpy())
print("C :" , C.numpy())
print("D :" , D.numpy())
print("loss :", loss.numpy())
```
